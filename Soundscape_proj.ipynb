```python
import joblib

# Load the trained model
model = joblib.load("genre_classification_model.pkl")  # Replace with the actual file path
```

```python
# Load the scaler
scaler = joblib.load("scaler.pkl")
```

```python
import numpy as np
import librosa
import pandas as pd
import json

# Load selected features from JSON
with open("selected_features.json", "r") as f:
    selected_features = json.load(f)

def extract_features(audio_path):
    # Load audio file
    y, sr = librosa.load(audio_path, mono=True, duration=30)
    
    # Compute required features
    features = {
        "chroma_stft_mean": np.mean(librosa.feature.chroma_stft(y=y, sr=sr)),
        "chroma_stft_var": np.var(librosa.feature.chroma_stft(y=y, sr=sr)),
        "length": len(y),  
        "rms_mean": np.mean(librosa.feature.rms(y=y)),
        "rms_var": np.var(librosa.feature.rms(y=y)),
        "spectral_bandwidth_mean": np.mean(librosa.feature.spectral_bandwidth(y=y, sr=sr)),
        "spectral_centroid_mean": np.mean(librosa.feature.spectral_centroid(y=y, sr=sr)),
        "spectral_centroid_var": np.var(librosa.feature.spectral_centroid(y=y, sr=sr)),
        "rolloff_mean": np.mean(librosa.feature.spectral_rolloff(y=y, sr=sr)),
        "rolloff_var": np.var(librosa.feature.spectral_rolloff(y=y, sr=sr)),
        "zero_crossing_rate_mean": np.mean(librosa.feature.zero_crossing_rate(y))
    }
    
    # Compute MFCCs 1-17
    mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=17)
    for i in range(1, 18):
        features[f"mfcc{i}_mean"] = np.mean(mfccs[i - 1])
        features[f"mfcc{i}_var"] = np.var(mfccs[i - 1])

    # Harmony and Perceived Harmonics
    harm, perceptr = librosa.effects.harmonic(y), librosa.effects.percussive(y)
    features["harmony_mean"] = np.mean(harm)
    features["harmony_var"] = np.var(harm)
    features["perceptr_mean"] = np.mean(perceptr)
    features["perceptr_var"] = np.var(perceptr)

    # Convert to DataFrame
    df = pd.DataFrame([features])

    # Ensure only selected features are kept
    df = df[list(selected_features)]  
    
    return df  #  Return the dataframe properly

#  Call the function outside
df = extract_features("hiphop.00003.wav")  # Replace with your actual file
print(df)  # This will display the extracted features
```

```python
# Extract features from the audio file
df = extract_features("hiphop.00003.wav")  # Replace with your actual file

# Scale the extracted features
df_scaled = scaler.transform(df)
```

```python
# Make a prediction using the trained model
prediction = model.predict(df_scaled)

# If you used LabelEncoder during training, convert it back to genre name
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
encoder.classes_ = np.array(['blues', 'classical', 'country', 'disco', 'hiphop', 
                             'jazz', 'metal', 'pop', 'reggae', 'rock'])  # Ensure this matches your training

predicted_genre = encoder.inverse_transform([prediction[0]])

print(" Predicted Genre:", predicted_genre[0])
```
